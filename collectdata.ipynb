```python
import requests
from bs4 import BeautifulSoup
import re


new_list = []
url = "https://news.ltn.com.tw/list/breakingnews/popular"


user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
headers = {"User-Agent":user_agent}

res = requests.get(url, headers=headers)
html = BeautifulSoup(res.text, 'html.parser')
#print(res.text)

storys = html.find_all("li",{"data-page":"1"})
for i in storys:

    try:
        story_title = i.a["title"]
        story_url =i.a["href"].replace(" ","")
        #story_url = "https://money.udn.com"+i.a["href"].replace(" ","")
        print('title: {},\n link: {}'.format(story_title, story_url))
        new_list.append(story_url)
    except:
        pass
```

```python
from time import sleep
import os
from bs4 import BeautifulSoup
from urllib.request import urlretrieve

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

        
def save_news(url):
    
    user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
    headers = {"User-Agent":user_agent} 
    response = requests.get(url, headers=headers)
    sleep(1)
    
    html = BeautifulSoup(response.text, "html.parser")
    try:
        title = html.find("div",{"class":"whitecon"}).h1.text
        content_str = html.find("div",{"class":"text boxTitle boxText"})
        content = content_str.find_all("p")[:-2]
        save_path = "./news/"+title.replace(':','')
        createFolder(save_path)
        with open(save_path+'/contents.txt', 'w', encoding='utf-8') as f:  
            print('new content: ')
            for i in content:
                print(i.text)
                f.write(i.text)
    except:
        pass

        
        

for new_url in new_list:
    print(new_url)
    save_news(new_url)
```
