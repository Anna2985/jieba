```python
import jieba
import os

source_file_path = "./news"
file_list = os.listdir(source_file_path)

all_airticle_string = ""
for each_article in file_list:
    article_path = source_file_path+"/"+each_article+"/contents.txt"
    with open (article_path,"r",encoding = "utf-8") as f:
        tmp_article_string = f.read()
        all_airticle_string += tmp_article_string + "\n"
    
print(all_airticle_string)
    
```

```python
import jieba
s_list = jieba.cut(all_airticle_string)
word_count = {}
for i in s_list:
    if i in word_count:
        word_count[i] += 1
    else:
        word_count[i] = 1
```
